```python
# Replace null values with zero using fillna
df_zero = df
df_zero = df.fillna(0)
df_zero.isnull().sum()
```

```python
# Replace null values with mean using fillna
df_mean = df
df_mean['HDI for year'] = df['HDI for year'].fillna(df['HDI for year'].mean())
df_mean.isnull().sum()
```

```python
# Replace null values with median using fillna
df_median = df
df_median['HDI for year'] = df['HDI for year'].fillna(df['HDI for year'].median())
df_median.isnull().sum()
```

```python
# Replace null values with mode using fillna
df_mode = df
df_mode['HDI for year'] = df['HDI for year'].fillna(df['HDI for year'].mode())
df_mode.isnull().sum()
```

```python
# implementation of the Tukey's algorithm
# values that look different from the other values in the data
# Box plots are a visual method to identify outliers
def findOutliers(dataset):
    outliers = []
    Q2 = dataset.median()
    Q1 = dataset.describe()[4]
    Q3 = dataset.describe()[6]
    
    # Q1 = dataset.quantile(0.25)
    # Q3 = dataset.quantile(0.75)
    
    # 
    
    IQR = Q3 - Q1
    lf = Q1 - 1.5*IQR # lf = lower fence
    uf = Q3 + 1.5*IQR # Uf = Upper fence
    for x in dataset:
        if x < lf or x > uf:
            outliers.append(x)
    return outliers
findOutliers(df['gdp_per_capita ($)'])
```
